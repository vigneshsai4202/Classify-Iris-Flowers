```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('/content/Iris_DT.csv')
x=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
y=['Species']
X=data[x]
Y=data[y]

data

data.info()

from sklearn import tree
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']],data['Species'],test_size=0.2)

from sklearn.metrics import confusion_matrix,classification_report

def modelpred (model, x_train,y_train, x_test):
  model=model
  model.fit(x_train,y_train)
  y_pred=model.predict(x_test)

  plt.figure(figsize=(10,15))
  tree.plot_tree(model,filled=True)
  plt.show()

  print(confusion_matrix(y_test,y_pred))
  print(classification_report(y_test,y_pred))
ID3=tree.DecisionTreeClassifier(criterion="entropy")
modelpred(ID3,x_train,y_train, x_test)

```
